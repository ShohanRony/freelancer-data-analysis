## Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot styles using Seaborn directly
sns.set_style("whitegrid")  # Use Seaborn's whitegrid style
sns.set_palette("Set2")  # Set color palette
plt.rcParams['figure.figsize'] = (12, 8)  # Set figure size
plt.rcParams['font.size'] = 12  # Set font size
df = pd.read_csv('/kaggle/input/freelancer-earnings-and-job-trends/freelancer_earnings_bd.csv')
```

## First look at the data

```python
print("Dataset shape:", df.shape)
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

## Check data types and missing values

```python
print("\nData information:")
print(df.info())
```

## Statistical summary of numerical columns

```python
print("\nStatistical summary of numerical columns:")
print(df.describe())
```

##  Check for missing values

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

## Check for duplicates

```python
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
```

## Explore categorical variables

```python
print("\nUnique values in categorical columns:")
categorical_cols = ['Job_Category', 'Platform', 'Experience_Level', 'Client_Region', 
                   'Payment_Method', 'Project_Type']
for col in categorical_cols:
    print(f"\n{col}: {df[col].nunique()} unique values")
    print(df[col].value_counts())
```

## Distribution of numerical variables


```python
numerical_cols = ['Job_Completed', 'Earnings_USD', 'Hourly_Rate', 'Job_Success_Rate', 
                  'Client_Rating', 'Job_Duration_Days', 'Rehire_Rate', 'Marketing_Spend']
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[col].dropna(), kde=True)  # Drop NaN values before plotting
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.savefig('numerical_distributions.png')  # Save the plot
plt.show()
```

## Correlation analysis

```python
plt.figure(figsize=(14, 10))
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.savefig('correlation_matrix.png')
plt.show()
```

## Relationship between categorical and numerical variable

## Earnings by Job Category

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Job_Category', y='Earnings_USD', data=df)
plt.title('Earnings by Job Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('earnings_by_category.png')
plt.show()
```

## Earnings by Experience Level

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Experience_Level', y='Earnings_USD', data=df)
plt.title('Earnings by Experience Level')
plt.tight_layout()
plt.savefig('earnings_by_experience.png')
plt.show()
```

## Relationship between Job Duration and Earnings

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Job_Duration_Days', y='Earnings_USD', hue='Project_Type', data=df)
plt.title('Relationship between Job Duration and Earnings')
plt.tight_layout()
plt.savefig('duration_vs_earnings.png')
plt.show()
```

## Average earnings by region

```python
plt.figure(figsize=(14, 8))
region_earnings = df.groupby('Client_Region')['Earnings_USD'].mean().sort_values(ascending=False)
sns.barplot(x=region_earnings.index, y=region_earnings.values)
plt.title('Average Earnings by Client Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('earnings_by_region.png')
plt.show()
```

## Job completion count by experience level and job category

```python
plt.figure(figsize=(16, 10))
completion_by_exp_cat = df.groupby(['Experience_Level', 'Job_Category'])['Job_Completed'].sum().unstack()
completion_by_exp_cat.plot(kind='bar')
plt.title('Job Completion Count by Experience Level and Job Category')
plt.ylabel('Number of Jobs Completed')
plt.legend(title='Job Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('completion_by_exp_cat.png')
plt.show()
```

## Client rating distribution by platform

```python
plt.figure(figsize=(14, 8))
sns.violinplot(x='Platform', y='Client_Rating', data=df)
plt.title('Client Rating Distribution by Platform')
plt.tight_layout()
plt.savefig('rating_by_platform.png')
plt.show()
```

## Relationship between marketing spend and earnings

```python
plt.figure(figsize=(12, 7))
sns.regplot(x='Marketing_Spend', y='Earnings_USD', data=df, scatter_kws={'alpha':0.5})
plt.title('Relationship between Marketing Spend and Earnings')
plt.tight_layout()
plt.savefig('marketing_vs_earnings.png')
plt.show()
```

## Top earning job categories

```python
top_categories = df.groupby('Job_Category')['Earnings_USD'].mean().sort_values(ascending=False)
plt.figure(figsize=(14, 8))
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title('Average Earnings by Job Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_earning_categories.png')
plt.show()
```

## Success rate by experience level

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Experience_Level', y='Job_Success_Rate', data=df)
plt.title('Job Success Rate by Experience Level')
plt.tight_layout()
plt.savefig('success_by_experience.png')
plt.show()
```

## Rehire rate analysis

```python
plt.figure(figsize=(14, 8))
rehire_by_category = df.groupby('Job_Category')['Rehire_Rate'].mean().sort_values(ascending=False)
sns.barplot(x=rehire_by_category.index, y=rehire_by_category.values)
plt.title('Average Rehire Rate by Job Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('rehire_by_category.png')
plt.show()
```

## Summary insights

```python
print(f"1. The dataset contains {df.shape[0]} freelancer records with {df.shape[1]} features.")
print(f"2. Most popular job category: {df['Job_Category'].value_counts().idxmax()}")
print(f"3. Most used platform: {df['Platform'].value_counts().idxmax()}")
print(f"4. Average earnings: ${df['Earnings_USD'].mean():.2f}")
print(f"5. Highest average earnings by job category: {top_categories.idxmax()} (${top_categories.max():.2f})")
print(f"6. Average job success rate: {df['Job_Success_Rate'].mean():.2f}%")
print(f"7. Correlation between earnings and job success rate: {df['Earnings_USD'].corr(df['Job_Success_Rate']):.2f}")
print(f"8. Correlation between marketing spend and earnings: {df['Marketing_Spend'].corr(df['Earnings_USD']):.2f}")
```
